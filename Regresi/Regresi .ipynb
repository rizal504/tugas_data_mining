```python
# import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
# Memanggil Dataset
df = pd.read_csv('Customers2.csv')
```

```python
#Drop kolom customer ID, Profesion, work experience, familiy size dan Gender
df = df.drop(['CustomerID', 'Gender', 'Profession','Age','Work Experience', 'Family Size'], axis=1)
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
print (x)
```

```python
print(y)
```

```python
df.keys()
```

```python
df.shape
```

```python
#menampilkan isi sebagian dataset 
data = pd.DataFrame(df)
data.head()
```

```python
# spilit Fitting Simple Linier Training set dan testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
#melakukan Fitting simple Linier Register pada Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() 
regressor.fit(x_train, y_train)
```

```python
#visualisasi data
plt.scatter(data.Annual_Income, data.Spending_Score)
plt.xlabel("Annual_Income")
plt.ylabel("Spending_Score")
plt.title("Grafik Age vs Annual_Income")
plt.show
```

```python
#visualisai hasil prediksi pada training set
#ukuran plot
plt.figure(figsize=(10,8))

#Biru adalah data observasi 
plt.scatter(x_train, y_train, color = 'blue')

#Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color = 'red')

#Memberi Judul dan label
plt.title("Age terhadap Annual Income (Training Set)")
plt.xlabel("Annual_Income")
plt.ylabel("Spending_Score")

#menampilkan plot
plt.show()
```

```python
#visualisai hasil prediksi pada test set
#ukuran plot
plt.figure(figsize=(10,8))

#Biru adalah data observasi 
plt.scatter(x_test, y_test, color = 'blue')

#Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_test, regressor.predict(x_test), color = 'red')

#Memberi Judul dan label
plt.xlabel("Annual_Income")
plt.ylabel("Spending_Score")
plt.title("Grafik Age vs Annual_Income")

#menampilkan plot
plt.show()
```
