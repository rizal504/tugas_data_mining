```python
#import Library
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
#membaca dataset
df = pd.read_csv("Customers.csv")
```

```python
#Menampilkan 5 data teratas
df.head()
```

```python
#Drop kolom customer ID, Profesion, work experience, familiy size dan Gender
df = df.drop(['CustomerID', 'Gender', 'Profession', 'Work Experience', 'Family Size'], axis=1)
```

```python
#Menampilkan 5 data teratas
df.head()
```

```python
# buat array 2D ke dalam variabel x. Data ini yang akan kita klaster.
x = np.array(df)
```

```python
# Menampilkan variabel x yang telah berisi array numpy
print(x)
```

```python
# lakukan hierachical klastering
Z = linkage(x)
```

```python
# plot dendogram
plt.figure(figsize=(25,15))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
