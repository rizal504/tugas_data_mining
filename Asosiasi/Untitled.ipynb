```python
# mengimport Library
import numpy as np
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#import data csv
df = pd.read_csv("Customers.csv")
df.head()
```

```python
#mengekplor kolom data
df.columns
```

```python
#mengeksplor berbagai  macam profesi
df.Profession.unique()
```

```python
#membershkan data 
#menghapus baris tanpa yers experience
df.dropna(axis=0, subset =['Work_Experience'], inplace= True)
df['Work_Experience'] = df['Work_Experience'].astype('str')

df = df[~df['Work_Experience'].str.contains('0')]
```

```python
#Memisahkan data sesuai dengan profesi
#profesi dilakukan Engineer
data_engineer = (df[df['Profession'] == "Engineer"]
                 .groupby(['Annual_Income', 'Spending_Score'])['Profession']
                 .count().unstack().reset_index().fillna(0)
                 .set_index('Annual_Income'))
data_engineer
```

```python
#mendefinisikan fungsi "Hot Encoding" untuk membuat dta sesuai untuk library yang bersangkutan
def hot_encode(x):
    if (x<=0):
        return 0
    if(x>=1):
        return 1
```

```python
engineer_encode = data_engineer.applymap(hot_encode)
data_engineer = engineer_encode
data_engineer
```

```python
#membangun model dan menganalisis hasilnya

# Menjalankan algoritma Apriori
frq_items = apriori(data_engineer, min_support= 0.005, use_colnames=True)

# Mengumpulkan aturan yang disimpulkan dalam DataFrame
rules = association_rules(frq_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, True])

# Mencetak hasil
print(rules.head())
```
